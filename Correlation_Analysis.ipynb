```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python
dataset = sns.load_dataset("mpg")
dataset.head()
```

```python
sns.scatterplot(data = dataset,
                x = "weight",
                y = "mpg")
plt.show()
```

```python
sns.scatterplot(data = dataset,
                x = "acceleration",
                y = "mpg")
plt.show()
```

```python
sns.scatterplot(data = dataset,
                x = "displacement",
                y = "mpg")
plt.show()
```

```python
correlation = dataset.corr()
correlation
```

```python
correlation = dataset.select_dtypes(include=np.number).corr()
correlation
```
